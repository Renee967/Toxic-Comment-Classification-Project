Bidirectional LSTM Model

```

from keras.layers import Dense, Input, LSTM, Bidirectional, Embedding, Dropout
from keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.layers import GlobalMaxPool1D
from keras.optimizers import Adam
from keras.models import Model
import tensorflow as tf
import numpy as np
import random
import time
import matplotlib.pyplot as plt

# fix seeds for re-producability

random.seed(7)
np.random.seed(7)
tf.random.set_seed(7)  # Use tf.random.set_seed in newer TensorFlow versions

def f1_score_metric(actual, predicted):
  predicted = tf.cast(predicted >= 0.5, tf.int32)  # Use tf.cast to convert to integers
  actual = tf.cast(actual, tf.int32)

  TP = tf.math.count_nonzero(predicted * actual)  # Use tf.math.count_nonzero
  TN = tf.math.count_nonzero((predicted - 1) * (actual - 1))
  FP = tf.math.count_nonzero(predicted * (actual - 1))
  FN = tf.math.count_nonzero((predicted - 1) * actual)

  precision = tf.divide(TP, TP + FP)
  recall = tf.divide(TP, TP + FN)

  f1 = tf.divide(2 * precision * recall, precision + recall)

  return f1
start_time = time.time()

TRAIN_DATA_FILE = "/content/drive/My Drive/toxic-comments-classification/toxic-nontoxic data/training-data-tokenized.csv"
VALIDATION_DATA_FILE = "/content/drive/My Drive/toxic-comments-classification/toxic-nontoxic data/validation-data-tokenized.csv"
EMBEDDING_FILE = "/content/drive/My Drive/toxic-comments-classification/toxic-nontoxic data/fasttext-embedding-matrix.csv"

# Parameters
max_comment_length = 150
vector_size = 300

# Read data
training_data = np.loadtxt(TRAIN_DATA_FILE, delimiter=',')
print(training_data.shape)

training_input = training_data[:, 0:max_comment_length]
print(training_input.shape)

training_labels = training_data[:, max_comment_length]
print(training_labels.shape)

embedding_matrix = np.loadtxt(EMBEDDING_FILE, delimiter=',')
print(embedding_matrix.shape)
vocab_count = embedding_matrix.shape[0]
print(vocab_count)

validation_data = np.loadtxt(VALIDATION_DATA_FILE, delimiter=',')

# Build model
input = Input(shape=(max_comment_length, ))

x = Embedding(input_dim=vocab_count, output_dim=vector_size, weights=[embedding_matrix], trainable=True)(input)

lstm_output = Bidirectional(LSTM(units=30, activation="tanh", return_sequences=True))(x)

pooling_output = GlobalMaxPool1D()(lstm_output)

dense1_output = Dense(80, activation="relu")(pooling_output)
dropout2_output = Dropout(0.2)(dense1_output)

dense2_output = Dense(80, activation="relu")(dropout2_output)
dropout3_output = Dropout(0.2)(dense2_output)

dense3_output = Dense(80, activation="relu")(dropout3_output)
dropout4_output = Dropout(0.2)(dense3_output)

output = Dense(1, activation="sigmoid")(dropout4_output)

# Compile Model
model = Model(inputs=input, outputs=output)
print(model.summary())

model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.005), metrics=['accuracy', f1_score_metric])

# Callbacks for early stopping and saving the best model
checkpoint = ModelCheckpoint('model-lstm.keras', monitor='val_f1_score_metric', # Changed the file extension to '.keras'
                             verbose=1, save_best_only=True, mode='max')

early_stopping = EarlyStopping(monitor='val_f1_score_metric', patience=5, mode='max')

callbacks_list = [checkpoint, early_stopping]

history = model.fit(training_input, training_labels, batch_size=2048, epochs=100,
                    callbacks=callbacks_list,
                    validation_data=(validation_data[:, 0:max_comment_length],
                                     validation_data[:,max_comment_length]))

# Plot F1 score
plt.plot(history.history['f1_score_metric'], label='Train F1 Score')
plt.plot(history.history['val_f1_score_metric'], label='Validation F1 Score')
plt.title('Model F1 Score')
plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.legend()
plt.show()

print("--- %s seconds ---" % (time.time() - start_time))
```

```
from sklearn.metrics import roc_auc_score
from keras.models import load_model
import matplotlib.pyplot as plt
from sklearn.metrics import auc, roc_curve, confusion_matrix, f1_score
import tensorflow as tf
import numpy as np
import random

# fix seeds for re-producability
random.seed(7)
np.random.seed(7)
tf.random.set_seed(7)  # Use tf.random.set_seed in newer TensorFlow versions

def f1_score_metric(actual, predicted):
  predicted = tf.cast(predicted >= 0.5, tf.int32)  # Use tf.cast to convert to integers
  actual = tf.cast(actual, tf.int32)

  TP = tf.math.count_nonzero(predicted * actual)  # Use tf.math.count_nonzero
  TN = tf.math.count_nonzero((predicted - 1) * (actual - 1))
  FP = tf.math.count_nonzero(predicted * (actual - 1))
  FN = tf.math.count_nonzero((predicted - 1) * actual)

  precision = tf.divide(TP, TP + FP)
  recall = tf.divide(TP, TP + FN)

  f1 = tf.divide(2 * precision * recall, precision + recall)

  return f1

VALIDATION_DATA_FILE = "/content/drive/My Drive/toxic-comments-classification/toxic-nontoxic data/validation-data-tokenized.csv"

# Read data
validation_data = np.loadtxt(VALIDATION_DATA_FILE, delimiter=',')
print(validation_data.shape)

validation_input = validation_data[:, 0:150]
print(validation_input.shape)

validation_labels = validation_data[:, 150]
print(validation_labels.shape)
# load model
model = load_model ("/content/drive/My Drive/toxic-comments-classification/toxic-nontoxic data/model-ensemble-0.828.h5", custom_objects={'f1_score_metric': f1_score_metric})
# model = load_model('drive/Deep Learning in NLP /toxic nontoxic**/model-ensemble-0.828.h5', custom_objects={'f1_score_metric': f1_score_metric})
# model = load_model('model-cnn.h5', custom_objects={'f1_score_metric': f1_score_metric})
# model = load_model('drive/toxic nontoxic**/model-lstm.h5', custom_objects={'f1_score_metric': f1_score_metric})

y_pred = model.predict(validation_input)

# Compute ROC
roc_auc = roc_auc_score(y_true=validation_labels, y_score=y_pred, average=None, sample_weight=None)
print('roc auc 1', roc_auc)

# pos_label is the positive label
fpr, tpr, thresholds = roc_curve(y_true=validation_labels, y_score=y_pred, pos_label=1)
roc_auc = auc(fpr, tpr)
print('roc auc 2', roc_auc)

y_pred[y_pred >= 0.5] = 1
y_pred[y_pred < 0.5] = 0

# Confusion matrix
tn, fp, fn, tp = confusion_matrix(y_true=validation_labels, y_pred=y_pred).ravel()
print(tn, fp, fn, tp)

accuracy = (tn + tp) / (tn+tp+fn+fp)
print('accuracy = ', accuracy)

# F1 score
f1 = f1_score(y_true=validation_labels, y_pred=y_pred, average=None)
print(f1)

weighted_f1 = f1_score(y_true=validation_labels, y_pred=y_pred, average='weighted')
print(weighted_f1)
```

**CNN MODEL**

```
from keras.layers import Dense, Input, Conv2D, Embedding, Dropout, Reshape, GlobalMaxPool2D
from keras.models import Model
from keras.optimizers import Adam

def calculate_vocab_size(text_data):
  """Calculates the vocabulary size of a given text dataset.

  Args:
    text_data: The text data as a list of strings.

  Returns:
    The size of the vocabulary.
  """

  # Replace this with your actual implementation to count unique words
  unique_words = set()
  for text in text_data:
    words = text.split()
    unique_words.update(words)
  return len(unique_words)

# Example usage:
your_text_data = ["This is a sample sentence.", "Another sentence with different words."]
vocab_size = calculate_vocab_size(your_text_data)

def create_cnn_model(vocab_size, embedding_dim, max_comment_length, embedding_matrix):
    input_layer = Input(shape=(max_comment_length,))
    x = Embedding(input_dim=embedding_matrix.shape[0],  # Use actual vocabulary size from matrix
                 output_dim=embedding_dim, weights=[embedding_matrix], trainable=True)(input_layer)
    x_reshaped = Reshape((max_comment_length, embedding_dim, 1))(x)
    cnn_output = Conv2D(filters=300, kernel_size=(4, embedding_dim), activation='relu', data_format="channels_last")(x_reshaped)
    pooling_output = GlobalMaxPool2D()(cnn_output)
    dense_output = Dense(50, activation="relu")(pooling_output)
    dropout_output = Dropout(0.2)(dense_output)
    output_layer = Dense(1, activation="sigmoid")(dropout_output)

    model = Model(inputs=input_layer, outputs=output_layer)
    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0002), metrics=['accuracy'])
    return model

# Assuming you have 'embedding_dim', 'max_comment_length' defined somewhere
# Replace these with the actual values you are using
embedding_dim = 300
max_comment_length = 150

# Load the embedding matrix (make sure the file path is correct)
embedding_matrix = np.loadtxt("/content/drive/My Drive/toxic-comments-classification/toxic-nontoxic data/fasttext-embedding-matrix.csv", delimiter=',') # Load embedding matrix

# Create and train the CNN model, use the calculated vocab_size
cnn_model = create_cnn_model(vocab_size, embedding_dim, max_comment_length, embedding_matrix)
cnn_model.fit(training_input, training_labels, batch_size=1024, epochs=10, validation_split=0.1)
```

**LSTM MODEL**

```
from keras.layers import LSTM

def create_lstm_model(vocab_size, embedding_dim, max_comment_length, embedding_matrix):
    input_layer = Input(shape=(max_comment_length,))
    x = Embedding(input_dim=embedding_matrix.shape[0],  # Use actual vocabulary size from matrix
                 output_dim=embedding_dim, weights=[embedding_matrix], trainable=True)(input_layer)
    lstm_output = LSTM(128)(x)
    dense_output = Dense(50, activation="relu")(lstm_output)
    dropout_output = Dropout(0.2)(dense_output)
    output_layer = Dense(1, activation="sigmoid")(dropout_output)

    model = Model(inputs=input_layer, outputs=output_layer)
    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0002), metrics=['accuracy'])
    return model

# Create and train the LSTM model
lstm_model = create_lstm_model(vocab_size, embedding_dim, max_comment_length, embedding_matrix)
lstm_model.fit(training_input, training_labels, batch_size=1024, epochs=10, validation_split=0.1)
```

**GRU MODEL**

```
from keras.layers import GRU

def create_gru_model(vocab_size, embedding_dim, max_comment_length, embedding_matrix):
    input_layer = Input(shape=(max_comment_length,))
    x = Embedding(input_dim=embedding_matrix.shape[0],  # Use actual vocabulary size from matrix
                 output_dim=embedding_dim, weights=[embedding_matrix], trainable=True)(input_layer)
    gru_output = GRU(128)(x)
    dense_output = Dense(50, activation="relu")(gru_output)
    dropout_output = Dropout(0.2)(dense_output)
    output_layer = Dense(1, activation="sigmoid")(dropout_output)

    model = Model(inputs=input_layer, outputs=output_layer)
    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0002), metrics=['accuracy'])
    return model

# Create and train the GRU model
gru_model = create_gru_model(vocab_size, embedding_dim, max_comment_length, embedding_matrix)
gru_model.fit(training_input, training_labels, batch_size=1024, epochs=10, validation_split=0.1)
```
